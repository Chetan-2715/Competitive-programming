```python
from array import array

# Creating an integer array
arr = array('i', [10, 20, 30, 40, 50])
print("Original Array:", arr)
```

```python
# Insert element at index 2
arr.insert(2, 99)
print("After insert at index 2:", arr)
```

```python
# Append element
arr.append(60)
print("After append:", arr)
```

```python
# Remove element (removes first occurrence)
arr.remove(40)
print("After remove 40:", arr)
```

```python
# Pop last element
removed = arr.pop()
print("Popped:", removed)
print("After pop:", arr)

# Pop at index 1
removed2 = arr.pop(1)
print("Popped index 1:", removed2)
print("After pop at index 1:", arr)
```

```python
# Reverse array
arr.reverse()
print("Reversed Array:", arr)

# Count occurrences of 10
print("Count of 10:", arr.count(10))

# Index of 30 (the first occurrence)
print("Index of 30:", arr.index(30))
```

```python
# Normal slicing
print("arr[1:4] =", arr[1:4])
print("arr[:3] =", arr[:3])
print("arr[2:] =", arr[2:])
print("arr[::-1] =", arr[::-1])  # Reverse using slicing
print("Full slice arr[:] =", arr[:])
```

```python
arr1 = array('i', [1, 2, 3, 4, 5])
arr2 = array('i', [10, 20, 30, 40])

# Concatenating by converting to list
result = array('i', arr1.tolist() + arr2.tolist())
print("Concatenated Array:", result)
```

```python
int_arr = array('i', [1, 2, 3, 4, 5])

# Creating float array
float_arr = array('f', [float(x) for x in int_arr])
print("Integer Array:", int_arr)
print("Float Array:", float_arr)
```
