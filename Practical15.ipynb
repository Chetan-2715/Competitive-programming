```python
def and_product(a, b):
    shift = 0
    while a != b:
        a >>= 1
        b >>= 1
        shift += 1
    return a << shift
```

```python
pairs = [(12,15), (2,3), (8,13)]

for a, b in pairs:
    print(f"AND product of [{a}, {b}] =", and_product(a, b))
```

```python
def great_xor(x):
    count = 0
    bit = 1

    while bit < x:
        if (x & bit) == 0:
            count += bit
        bit <<= 1

    return count
```

```python
examples = [2, 5, 10]

for x in examples:
    print(f"Great XOR for {x} =", great_xor(x))
```
