```python
arr = [10,50,30,70,80,60,20,90,40]
search = 20

position = -1

for i in range(len(arr)):
    if arr[i] == search:
        position = i
        break

print("Array:", arr)
print("Element to search:", search)

if position != -1:
    print(f"Element {search} found at position {position}")
else:
    print("Element not found.")
```

```python
arr = sorted([10,50,30,70,80,60,20,90,40])
search = 90

left, right = 0, len(arr)-1
found_at = -1

while left <= right:
    mid = (left + right) // 2
    if arr[mid] == search:
        found_at = mid
        break
    elif search < arr[mid]:
        right = mid - 1
    else:
        left = mid + 1

print("Sorted Array:", arr)
print("Search:", search)
print("Found at index:", found_at)
```

```python
arr = [12,11,13,5,6,10]

for i in range(len(arr)):
    min_pos = i
    for j in range(i+1, len(arr)):
        if arr[j] < arr[min_pos]:
            min_pos = j
    
    arr[i], arr[min_pos] = arr[min_pos], arr[i]

print("Selection Sort Output:", arr)
```

```python
arr = [12,11,13,5,6,10]

for i in range(1, len(arr)):
    key = arr[i]
    j = i - 1
    while j >= 0 and arr[j] > key:
        arr[j+1] = arr[j]
        j -= 1
    arr[j+1] = key

print("Insertion Sort Output:", arr)
```

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr)//2
        left = arr[:mid]
        right = arr[mid:]

        merge_sort(left)
        merge_sort(right)

        i = j = k = 0

        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                arr[k] = left[i]
                i += 1
            else:
                arr[k] = right[j]
                j += 1
            k += 1

        while i < len(left):
            arr[k] = left[i]
            i += 1
            k += 1
        while j < len(right):
            arr[k] = right[j]
            j += 1
            k += 1

arr = [12,11,13,5,6,10]
merge_sort(arr)
print("Merge Sort Output:", arr)
```

```python
def heapify(arr, n, i):
    largest = i
    left = 2*i + 1
    right = 2*i + 2

    if left < n and arr[left] > arr[largest]:
        largest = left

    if right < n and arr[right] > arr[largest]:
        largest = right

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def heap_sort(arr):
    n = len(arr)

    for i in range(n//2 - 1, -1, -1):
        heapify(arr, n, i)

    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)

arr = [12,11,13,5,6,10]
heap_sort(arr)
print("Heap Sort Output:", arr)
```

```python
arr = [12,11,13,5,6,10]

for i in range(len(arr)):
    for j in range(0, len(arr)-i-1):
        if arr[j] > arr[j+1]:
            arr[j], arr[j+1] = arr[j+1], arr[j]

print("Bubble Sort Output:", arr)
```

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr)//2]
    left = [x for x in arr if x < pivot]
    mid = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + mid + quick_sort(right)

arr = [12,11,13,5,6,10]
arr = quick_sort(arr)
print("Quick Sort Output:", arr)
```
