```python
def fibonacci_dp(n):
    dp = [0] * (n+1)
    dp[0] = 0
    dp[1] = 1

    for i in range(2, n+1):
        dp[i] = dp[i-1] + dp[i-2]

    return dp[n]

print("Fibonacci(9) =", fibonacci_dp(9))
```

```python
def modified_fibonacci(t1, t2, n):
    dp = [0] * (n+1)
    dp[1] = t1
    dp[2] = t2

    for i in range(3, n+1):
        dp[i] = dp[i-2] + dp[i-1]**2

    return dp[n]


print("Modified Fibonacci (0,1,5) =", modified_fibonacci(0, 1, 5))
```

```python
def substring_sum(s):
    MOD = 10**9 + 7
    total = 0
    prev = 0

    for i, ch in enumerate(s):
        digit = int(ch)
        curr = (prev * 10 + digit * (i+1)) % MOD
        total = (total + curr) % MOD
        prev = curr

    return total


print("Substring Sum of '123' =", substring_sum("123"))
```
