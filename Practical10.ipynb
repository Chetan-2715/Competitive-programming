```python
# Stack using Python list
stack = []

# Push
stack.append(10)
stack.append(20)
stack.append(30)

print("Stack after pushing 10,20,30:", stack)

# Pop
popped = stack.pop()
print("Popped element:", popped)
print("Stack after pop:", stack)
```

```python
# Queue using Python list
queue = []

# Enqueue (push)
queue.append(5)
queue.append(15)
queue.append(25)

print("Queue after enqueuing 5,15,25:", queue)

# Dequeue (pop front)
removed = queue.pop(0)
print("Dequeued element:", removed)
print("Queue after dequeue:", queue)
```

```python
from collections import deque

dq = deque()

# Push elements
dq.append(100)
dq.appendleft(50)
dq.append(150)

print("Deque after pushes:", dq)

# Pop elements
dq.pop()
dq.popleft()

print("Deque after pops:", dq)
```

```python
class TwoStackQueue:
    def __init__(self):
        self.stack1 = []
        self.stack2 = []

    def enqueue(self, x):
        self.stack1.append(x)

    def dequeue(self):
        if not self.stack2:
            while self.stack1:
                self.stack2.append(self.stack1.pop())
        
        return self.stack2.pop() if self.stack2 else None


# Example
q = TwoStackQueue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)

print("Dequeue:", q.dequeue())  # should remove 1
print("Dequeue:", q.dequeue())  # should remove 2
```

```python
def two_stack_game(X, A, B):
    i = j = 0
    current_sum = 0
    count = 0

    # first take from A as long as possible
    while i < len(A) and current_sum + A[i] <= X:
        current_sum += A[i]
        i += 1
        count += 1

    # now try taking from B, removing from A if needed
    max_score = count

    while j < len(B):
        current_sum += B[j]
        j += 1

        while current_sum > X and i > 0:
            i -= 1
            current_sum -= A[i]
            count -= 1

        if current_sum > X:
            break

        max_score = max(max_score, i + j)

    return max_score


X = 10
A = [4,2,4,6,1]
B = [2,1,8,5]

print("Maximum Score:", two_stack_game(X, A, B))
```
