```python
def is_safe(board, row, col):
    # Check this column
    for i in range(row):
        if board[i][col] == 1:
            return False

    # Check upper-left diagonal
    i, j = row, col
    while i >= 0 and j >= 0:
        if board[i][j] == 1:
            return False
        i -= 1
        j -= 1

    # Check upper-right diagonal
    i, j = row, col
    while i >= 0 and j < 4:
        if board[i][j] == 1:
            return False
        i -= 1
        j += 1

    return True
```

```python
def solve_4_queens(board, row):
    if row == 4:
        return True   # solution complete

    for col in range(4):
        if is_safe(board, row, col):
            board[row][col] = 1   # place queen

            if solve_4_queens(board, row + 1):
                return True

            board[row][col] = 0   # backtrack

    return False
```

```python
board = [[0] * 4 for _ in range(4)]

if solve_4_queens(board, 0):
    print("4-Queens Solution:\n")
    for row in board:
        print(["Q" if x == 1 else "." for x in row])
else:
    print("No solution exists.")
```
