```python
# Graph represented as adjacency list
graph = {
    5: [3, 7],
    3: [2, 4],
    7: [8],
    2: [],
    4: [],
    8: []
}

print("Graph adjacency list:")
for node in graph:
    print(node, "->", graph[node])
```

```python
from collections import deque

def bfs(graph, start):
    visited = set()
    queue = deque([start])
    order = []

    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.add(node)
            order.append(node)
            for neighbor in graph[node]:
                queue.append(neighbor)

    return order
```

```python
def dfs(graph, start, visited=None, order=None):
    if visited is None:
        visited = set()
    if order is None:
        order = []

    visited.add(start)
    order.append(start)

    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited, order)

    return order
```

```python
print("BFS Traversal:", bfs(graph, 5))
print("DFS Traversal:", dfs(graph, 5))
```
