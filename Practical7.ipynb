```python
class BSTNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


def bst_insert(root, key):
    if root is None:
        return BSTNode(key)
    
    if key < root.data:
        root.left = bst_insert(root.left, key)
    else:
        root.right = bst_insert(root.right, key)

    return root
```

```python
def bst_search(root, key):
    if root is None:
        return False
    if root.data == key:
        return True
    if key < root.data:
        return bst_search(root.left, key)
    return bst_search(root.right, key)
```

```python
def find_min(node):
    while node.left:
        node = node.left
    return node

def bst_delete(root, key):
    if root is None:
        return None
    
    if key < root.data:
        root.left = bst_delete(root.left, key)
    elif key > root.data:
        root.right = bst_delete(root.right, key)
    else:
        # Case 1: no child
        if root.left is None and root.right is None:
            return None
        
        # Case 2: one child
        if root.left is None:
            return root.right
        if root.right is None:
            return root.left
        
        # Case 3: two children
        min_node = find_min(root.right)
        root.data = min_node.data
        root.right = bst_delete(root.right, min_node.data)

    return root
```

```python
root = None
values = [1, 2, 3, 4, 5]

for v in values:
    root = bst_insert(root, v)

print("BST Created with values:", values)
```

```python
print("Before insert:")
# (We'll print inorder later)
    
# Insert 10
root = bst_insert(root, 10)
print("Inserted 10")

# Delete 3
root = bst_delete(root, 3)
print("Deleted 3")
```

```python
def inorder(root):
    return inorder(root.left) + [root.data] + inorder(root.right) if root else []

def preorder(root):
    return [root.data] + preorder(root.left) + preorder(root.right) if root else []

def postorder(root):
    return postorder(root.left) + postorder(root.right) + [root.data] if root else []
```

```python
print("Inorder Traversal:", inorder(root))
print("Preorder Traversal:", preorder(root))
print("Postorder Traversal:", postorder(root))
```
