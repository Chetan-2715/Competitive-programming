```python
s = "Hello World 123"
print("Original String:", s)
```

```python
print("Uppercase:", s.upper())
print("Lowercase:", s.lower())
print("Replace 'World' → 'Python':", s.replace("World", "Python"))
print("Find 'World':", s.find("World"))
print("Split:", s.split())
print("Starts with 'Hello'?", s.startswith("Hello"))
print("Is numeric?", s.isnumeric())

# Index of character
print("Index of 'W':", s.index("W"))
```

```python
string = "ABCDCDC"
sub = "CDC"
count = 0

for i in range(len(string) - len(sub) + 1):
    if string[i:i+len(sub)] == sub:
        count += 1

print("String:", string)
print("Substring:", sub)
print("Occurrences:", count)
```

```python
s1 = "this is string"
s2 = "THIS IS STRING"

print("Swapcase 1:", s1.swapcase())
print("Swapcase 2:", s2.swapcase())
```

```python
tests = ["ab123", "abcgd", "1234"]

for t in tests:
    print("\nTesting:", t)

    if t.isalpha():
        print("String contains alphabets only.")
    elif t.isdigit():
        print("String contains digits only.")
    elif t.isalnum():
        print("String is alphanumeric.")
    else:
        print("String is not alphanumeric.")
```

```python
s = "abcdefg"
pos = 5          # 0-based indexing → 5th index = 6th character
new_char = "k"

modified = s[:pos] + new_char + s[pos+1:]
print("Original:", s)
print("Modified:", modified)
```

```python
text = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
width = 4

print("Wrapped Text:")
for i in range(0, len(text), width):
    print(text[i:i+width])
```
