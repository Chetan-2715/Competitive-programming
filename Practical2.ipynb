```python
# Example matrices (as given in manual)

mat1 = [
    [12, 7, 3],
    [4, 5, 6],
    [7, 8, 9]
]

mat2 = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print("Matrix 1:")
for row in mat1:
    print(row)

print("\nMatrix 2:")
for row in mat2:
    print(row)
```

```python
# Matrix addition (3x3)
result_add = []

for i in range(len(mat1)):
    temp = []
    for j in range(len(mat1[0])):
        temp.append(mat1[i][j] + mat2[i][j])
    result_add.append(temp)

print("Addition of Matrix 1 and Matrix 2:")
for row in result_add:
    print(row)
```

```python
matrix = [
    [1, 4],
    [2, 5],
    [3, 6]
]

print("Original Matrix:")
for row in matrix:
    print(row)
```

```python
# Transpose (convert rows â†’ columns)
transpose = []

for i in range(len(matrix[0])):       # number of columns
    row = []
    for j in range(len(matrix)):      # number of rows
        row.append(matrix[j][i])
    transpose.append(row)

print("Transpose of Matrix:")
for row in transpose:
    print(row)
```
